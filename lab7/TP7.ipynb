```python
#Exo D
#Question 1
def tf(word, textblob):
    return textblob.words.count(word)/len(textblob.words)
```

```python
def n_containing(word, textbloblist):
    compt = 0
    for i in textbloblist:
        if i.words.count(word) != 0:
            compt += 1
    return compt
```

```python
def idf(word, textbloblist):
    doc_len = 0
    word_compt = 0
    for i in textbloblist:
        doc_len += len(i.words)
        word_compt += i.words.count(word)
    return 1/(word_compt/doc_len)
```

```python
def tfidf(word, textblob, textbloblist):
    return tf(word, textblob)*idf(word, textbloblist)
```

```python
#Question 2
bloblist = []
```

```python
bloblist.append(TextBlob(wikipedia.summary("France")))
bloblist.append(TextBlob(wikipedia.summary("Elton John")))
bloblist.append(TextBlob(wikipedia.summary("Python")))
bloblist.append(TextBlob(wikipedia.summary("Fox")))
bloblist.append(TextBlob(wikipedia.summary("Isaac Newton")))
bloblist.append(TextBlob(wikipedia.summary("Zinedine Zidane")))
```

```python
stop_words = stopwords.words('english')
print(stop_words)
```

```python
#Question 3
bloblist_lower = [textblob.lower() for textblob in bloblist]
words_bag = []
stop_words = stopwords.words('english')
for textblob in bloblist_lower:
    for word in textblob.words:
        if word not in stop_words:
            words_bag.append(word)
```

```python
print(words_bag)
```

```python
#Question 4
for word in words_bag:
    print(word)
    for textblob in bloblist:
        print(tfidf(word, textblob, bloblist))
```
